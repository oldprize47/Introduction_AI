## Exercise 2 - Add new columns

### 1) Create a new column family_size as the sum of SibSp and Parch

```python
df['family_size'] = df['SibSp'] + df['Parch']
df[['family_size']]
```

### 2) Create a new column age_group that labels each passenger as:

```python
df['age_group'] = df['Age']
df['age_group'] = df['age_group'] .fillna(-1)
for i in range(len(df)):
    if df.iloc[i, 12] < 0:
        df.iloc[i, 12] = 'unknown' # 'unknown' if Age is missing
    elif df.iloc[i, 12] <= 12:
        df.iloc[i, 12] = 'child' # 'child' (Age ≤ 12)
    elif df.iloc[i, 12] < 60:
        df.iloc[i, 12] = 'adult' # 'adult' (12 < Age < 60)
    else:
        df.iloc[i, 12] = 'senior' # 'senior' (Age ≥ 60)
```

###  3) Convert the Sex column to numeric values (male → 0, female → 1) and store it as a new column Sex_bin


```python
df['Sex_bin'] = df['Sex'].apply(lambda x: 0 if x == 'male' else 1)
```

## Exercise 4 - Handling missing values

### 1) Print the number of missing values in the columns Age, Cabin, and Embarked

```python
print('Number of null values in the Age column =', len(df[df['Age'].isnull()]))
print('Number of null values in the Cabin column =', len(df[df['Cabin'].isnull()]))
print('Number of null values in the Embarked  column =', len(df[df['Embarked'].isnull()]))
```

### 2) Fill missing values in Age using the mean age per sex

```python
male_age_mean = df[df['Sex'] == 'male']['Age'].mean()
female_age_mean = df[df['Sex'] == 'female']['Age'].mean()

for i in range(len(df)):
    if df.iloc[i, 3] == 'male' and df.isnull().iloc[i, 4] == True:
        df.iloc[i, 4] = male_age_mean
    elif df.iloc[i, 3] == 'female' and df.isnull().iloc[i, 4] == True:
        df.iloc[i, 4] = female_age_mean

df[['Sex', 'Age']].head(30)
```

### 3) Drop the Cabin column

```python
df.drop(columns = ['Cabin'], inplace = True)
df.head(1)
```
